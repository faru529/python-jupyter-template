### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(11) 
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

try:
    # Check if categories and values have the same length
    if len(categories) != len(values):
        raise ValueError("Categories and values lists must have the same length.")
    
    # Check if labels and sizes have the same length
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes lists must have the same length.")
    
    # Check if sizes add up to a reasonable total (e.g., 100 for percentage-based pie chart)
    if sum(sizes) <= 0:
        raise ValueError("Sum of pie chart sizes must be greater than zero.")

    # Check if x and y have the same length for line and scatter plots
    if len(x) != len(y):
        raise ValueError("X and Y data for plots must have the same length.")

    # Create a 2x2 grid of subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # First subplot: Line plot
    axs[0, 0].plot(x, y, marker='o', color='b', linestyle='-')
    axs[0, 0].set_title('Line Plot')
    axs[0, 0].set_xlabel('X')
    axs[0, 0].set_ylabel('Y = X')

    # Second subplot: Bar plot
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title('Bar Plot')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    # Third subplot: Scatter plot
    axs[1, 0].scatter(x, y, color='orange', marker='s')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('X')
    axs[1, 0].set_ylabel('Y = X')

    # Fourth subplot: Pie chart
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
    axs[1, 1].set_title('Pie Chart')
    
    plt.tight_layout()
    plt.show()

except ValueError as e:
    print(f"Error: {e}")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
