### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
def plot_bar_chart(categories, values, colors):
    try:
        # Validate input types
        if not all(isinstance(i, list) for i in [categories, values, colors]):
            raise TypeError("Categories, values, and colors should be lists.")
        
        # Validate non-empty lists
        if not categories or not values or not colors:
            raise ValueError("Categories, values, and colors cannot be empty.")

        # Validate length match
        if len(categories) != len(values) or len(categories) != len(colors):
            raise ValueError("Categories, values, and colors must have the same length.")
        
        # Validate that values are numeric
        if not all(isinstance(v, (int, float)) for v in values):
            raise TypeError("All values must be numeric (int or float).")

        plt.bar(categories, values, color=colors)
        plt.xlabel("Categories")
        plt.ylabel("Values")
        plt.title("BAR PLOT")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

plot_bar_chart(['A', 'B', 'C', 'D'], [5, 7, 3, 9], ['Red', 'Green', 'Blue', 'Yellow'])
```
