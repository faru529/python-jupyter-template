### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
if not data:
    print("Error: The data array is empty. Please provide valid data.")
elif not all(isinstance(i, (int, float)) for i in data):
    print("Error: The data array contains non-numeric values. Please provide a valid numeric dataset.")
else:
    plt.hist(data, bins=5, color = 'red', edgecolor='black')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram of Data')
    plt.show()
```
